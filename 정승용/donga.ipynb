```python
import requests
from bs4 import BeautifulSoup

# URL 설정
url = "https://www.donga.com/news/Politics/List?ymd=20230601&m="

# 페이지 가져오기
response = requests.get(url)
html_content = response.text

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(html_content, "html.parser")

# 각 뉴스 링크 가져오기
news_links = soup.find_all("p", class_="tit")

for link in news_links:
    news_url = link.a["href"]  # 뉴스의 링크
    news_title = link.text.strip()  # 뉴스의 제목
    print("제목:", news_title)
    print("링크:", news_url)
    
    # 뉴스의 내용 가져오기
    news_response = requests.get(news_url)
    news_html_content = news_response.text
    news_soup = BeautifulSoup(news_html_content, "html.parser")
    
    # 뉴스의 본문 내용 가져오기
    news_content = news_soup.find("div", class_="article_txt").text.strip()
    print("내용:", news_content)
    print("="*50)
```
